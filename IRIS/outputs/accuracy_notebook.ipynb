```python
import os
import matplotlib.pyplot as plt

# Folder where accuracy files are stored
accuracy_folder = '../outputs/Accuracy'

# Empty dictionary to hold model names and their accuracies
accuracy_scores = {}

# Loop through all files in the accuracy folder
for filename in os.listdir(accuracy_folder):
    if filename.endswith('.txt'):
        model_name = filename.replace('_accuracy.txt', '').replace('_', ' ').title()
        file_path = os.path.join(accuracy_folder, filename)
        with open(file_path, 'r') as file:
            line = file.readline()
            try:
                # Extract accuracy float value from the line
                accuracy = float(line.strip().split()[-1])
                accuracy_scores[model_name] = accuracy
            except:
                print(f"Could not extract accuracy from {filename}")

# Sort by accuracy (optional)
accuracy_scores = dict(sorted(accuracy_scores.items(), key=lambda x: x[1], reverse=True))

# Plotting
plt.figure(figsize=(12, 6))
plt.bar(accuracy_scores.keys(), accuracy_scores.values(), color='skyblue')
plt.xticks(rotation=45, ha='right')
plt.ylim(0.8, 1.05)
plt.title("Accuracy Comparison of 10 ML Models")
plt.ylabel("Accuracy Score")
plt.tight_layout()

# Save and show
plt.savefig('../outputs/model_accuracy_comparison.png')
plt.show()
```
